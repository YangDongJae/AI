```python
from gensim.models.doc2vec import Doc2Vec, TaggedDocument
```

```python
tagged_data = [TaggedDocument(words=str(_d).split(), tags=[str(i)]) for i, _d in enumerate(df.eng_title)]

max_epochs = 10
vec_size = 128
alpha = 0.025

model = Doc2Vec(size = vec_size,
               alpha = alpha,
               min_alpha = 0.00025,
               min_count = 10,
               dm = 1)

model.build_vocab(tagged_data)

for epoch in range (10):
    
    print('iteration {0}'.format(epoch))
    
    model.train(tagged_data,
               total_examples = model.corpus_count,
               epochs = model.iter)
    
    model.alpha -= 0.0002
    model.min_alpha = model.alpha
    
print("model Trained")
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters = 10 , random_state = 0).fit(model.docvecs.doctag_syn0)
index_list = []

for k in range (len(kmeans.cluster_centers_)):
    print('Topic {}'.format(k))
    for i in model.docvecs.most_similar(positive=[kmeans.cluster_centers_[k]], topn = 10):
        print(i[0] , df['eng_title'][int(i[0])])
        print('\n')
```
